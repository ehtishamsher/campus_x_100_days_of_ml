```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LinearRegression
```

## using loop

```python
n_iterations = 5  # <- this is the value you'd change to "loop more"

df_current = df0.copy()

for it in range(n_iterations):
    df_prev = df_current.copy()
    df_current = df_prev.copy()

    # R&D Spend
    df_current.iloc[1,0] = np.nan
    X = df_current.iloc[[0,2,3,4],1:3]
    y = df_current.iloc[[0,2,3,4],0]
    lr = LinearRegression().fit(X, y)
    df_current.iloc[1,0] = lr.predict(df_current.iloc[1,1:].values.reshape(1,2))[0]

    # Administration
    df_current.iloc[3,1] = np.nan
    X = df_current.iloc[[0,1,2,4],[0,2]]
    y = df_current.iloc[[0,1,2,4],1]
    lr = LinearRegression().fit(X, y)
    df_current.iloc[3,1] = lr.predict(df_current.iloc[3,[0,2]].values.reshape(1,2))[0]

    # Marketing Spend
    df_current.iloc[4,-1] = np.nan
    X = df_current.iloc[0:4,0:2]
    y = df_current.iloc[0:4,-1]
    lr = LinearRegression().fit(X, y)
    df_current.iloc[4,-1] = lr.predict(df_current.iloc[4,0:2].values.reshape(1,2))[0]

    delta = (df_current - df_prev).abs().sum().sum()
    print(f"Iteration {it+1}, total change: {delta:.4f}")

    if delta < 0.01:  # convergence check, similar to sklearn's `tol`
        break
```

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.read_csv('covid_toy.csv')
```

```python
df.sample(10)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['has_covid']),df['has_covid'],test_size=0.2,random_state=42)
```

```python
df.isnull().sum()
```

```python
from sklearn.compose import ColumnTransformer
```

```python
tranformer = ColumnTransformer(transformers=[
    ('tnf1' , SimpleImputer(),['fever']),
    ('tnf2' , OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('tnf3', OneHotEncoder(sparse_output=False,drop='first'),['gender','city'])
],remainder='passthrough')
```

```python
tranformer.fit_transform(X_train).shape
```

```python
tranformer.transform(X_test).shape
```
